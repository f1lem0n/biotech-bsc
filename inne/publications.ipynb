```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
txt = open('data/analyze.txt', 'r')
txt = txt.readlines()
lines = []
for line in txt:
    lines.append(line.replace('\t', ', ').replace('\n', ''))
lines
csv = open('data/pub.csv', 'w')
for line in sorted(lines)[:57]:
    csv.write(line + '\n')
csv.close()
```

```python
df = pd.read_csv(
    'data/pub.csv',
    names=(
        'Rok publikacji',
        'Liczba publikacji',
        'Procent publikacji'
    )
)
df
```

```python
plt.figure(figsize=(8, 4), dpi=300, facecolor='#ffffff')
plt.bar(x=df['Rok publikacji'], height=df['Liczba publikacji'])
plt.ylabel('Liczba publikacji')
plt.xlabel('Rok publikacji')
plt.savefig('pub.png')
```
