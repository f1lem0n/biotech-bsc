```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(
    'data/pub.csv',
    names=['Rok', 'Liczba publikacji', '% publikacji']
)
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=300)
plt.style.use('seaborn')
ax.bar(
    x=df['Rok'],
    height=df['Liczba publikacji'],
    align='center',
    alpha=0.5,
    color='blue',
)
ax.grid(False)
ax.set_ylabel(
    'Liczba publikacji',
    labelpad=20,
    fontdict={
        'fontsize': 18
    }
)
ax.set_xlabel(
    'Rok',
    labelpad=20,
    fontdict={
        'fontsize': 18
    }
)
ax.set_ylim([0, 3000])
ax.grid(False)
ax.grid(axis='y', color='grey')
fig.savefig('../manuscript/figures/publications.png', bbox_inches='tight')
plt.show()
plt.close(fig)
```
